```python
import pandas as pd

# Load the data
file_path = '/Users/jitendrasingh/Downloads/Case Study 2_Office Supplies/Office Supplies Orders Case Study Data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data
```

```python
#1. Data Preparation
# Clean the data
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d/%m/%y')
data['Sales'] = data['Sales'].replace('[\$,]', '', regex=True).astype(float)
data['Profit'] = data['Profit'].replace('[\$,]', '', regex=True).astype(float)
data['Shipping Cost'] = data['Shipping Cost'].replace('[\$,]', '', regex=True).astype(float)
```

```python
#2. Descriptive Analysis
# Summary statistics
print(data.describe())

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Sales'], bins=30, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Profit distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Profit'], bins=30, kde=True)
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
#3. Sales Performance Analysis
# Top-performing products
top_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)

# Sales by region
sales_by_region = data.groupby('Region')['Sales'].sum()
print(sales_by_region)

# Visualization
plt.figure(figsize=(12, 8))
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(12, 8))
sales_by_region.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
#4. Profitability Analysis
# Profit by product
profit_by_product = data.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
print(profit_by_product)

# Profit margins by region
profit_margins_by_region = data.groupby('Region')['Profit'].sum() / data.groupby('Region')['Sales'].sum()
print(profit_margins_by_region)

# Visualization
plt.figure(figsize=(12, 8))
profit_by_product.plot(kind='bar')
plt.title('Top 10 Products by Profit')
plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.show()

plt.figure(figsize=(12, 8))
profit_margins_by_region.plot(kind='bar')
plt.title('Profit Margins by Region')
plt.xlabel('Region')
plt.ylabel('Profit Margin')
plt.show()
```

```python
#5. Shipping Efficiency Analysis
# Shipping cost by mode
shipping_cost_by_mode = data.groupby('Ship Mode')['Shipping Cost'].mean()
print(shipping_cost_by_mode)

# Profit by shipping mode
profit_by_shipping_mode = data.groupby('Ship Mode')['Profit'].sum()
print(profit_by_shipping_mode)

# Visualization
plt.figure(figsize=(12, 8))
shipping_cost_by_mode.plot(kind='bar')
plt.title('Average Shipping Cost by Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Average Shipping Cost')
plt.show()

plt.figure(figsize=(12, 8))
profit_by_shipping_mode.plot(kind='bar')
plt.title('Total Profit by Shipping Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Profit')
plt.show()
```

```python
#6. Customer Segmentation
# Customer segmentation
customer_segments = data.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum', 'Order ID': 'count'})
customer_segments.columns = ['Total Sales', 'Total Profit', 'Order Count']
print(customer_segments.head())

# Visualization
plt.figure(figsize=(12, 8))
sns.scatterplot(data=customer_segments, x='Total Sales', y='Total Profit', size='Order Count', hue='Order Count', palette='viridis', sizes=(20, 200))
plt.title('Customer Segmentation')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.show()
```
